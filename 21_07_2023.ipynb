```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.api.types import CategoricalDtype
```

```python
sns.set(rc={'figure.figsize':(6,4)})
```

```python
df_diamonds = sns.load_dataset('diamonds')
df_diamonds.head()
```

```python
df_diamonds.info()
```

```python
df_diamonds.columns
```

```python
df_diamonds["cut"].unique()
```

```python
df_diamonds["cut"].value_counts()
```

```python
df_diamonds["cut"].head()
```

```python
df_diamonds["cut"].value_counts().plot.barh().set_title("Class Frequencies of Cut Variable")
```

```python
df_diamonds["cut"].value_counts().plot.bar().set_title("Class Frequencies of Cut Variable")
```

```python
df_diamonds["cut"].index
```

```python
sns.barplot(x="cut", y=df_diamonds["cut"].index, data=df_diamonds, color="purple").set_title("Cut Density")
```

```python
sns.barplot(x="cut", y="price", hue="color", data=df_diamonds)
plt.title("Cut - Price - Color")
```

```python
sns.barplot(x="cut",
            y="price",
            hue="clarity",
            data=df_diamonds)
plt.title("Cut - Price - Clarity")
```

```python
sns.barplot(x="cut",
            y="price",
            hue="carat",
            data=df_diamonds)
plt.title("Cut - Price - Carat")
```

```python
df_diamonds.info()
```

```python
df_diamonds.groupby("cut").sum().plot.pie(y="price", autopct="%.1f%%", ylabel="", legend=False, figsize=(5,5))
```

```python
cols = ["x", "y", "z"]

for col in cols:
  plt.figure(figsize=(5,5))
  sns.scatterplot(x=col, y="price", data=df_diamonds)
```

```python
sns.histplot(df_diamonds["price"], kde=True) # bins= auto
```

```python
sns.histplot(df_diamonds["price"], kde=True, bins=40)
```

```python
sns.kdeplot(df_diamonds["price"])
```

```python
sns.jointplot(x="price", y=df_diamonds["carat"], data=df_diamonds)
```

```python
sns.catplot(x="cut", y="price", data=df_diamonds)
plt.title("Cut - Price")
```

```python
sns.catplot(x="cut", y="price", hue="color", kind="point", data=df_diamonds)
plt.title("Cut - Price")
```

```python
sns.kdeplot(df_diamonds["price"], shade=True)
```

```python
(sns.FacetGrid(df_diamonds, hue="cut", height=5, xlim=(0, 10000)).map(sns.kdeplot, "price", shade=True).add_legend())
```

```python
plt.title("Correlation Matrix")
sns.heatmap(df_diamonds.corr(), annot=True);
```
