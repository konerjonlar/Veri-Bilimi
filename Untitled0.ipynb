```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dosya_adı = "/content/AB_NYC_2019.csv"  # Yüklenen CSV dosyasının tam yolu
df = pd.read_csv(dosya_adı)
sns.set(rc={"figure.figsize":(20,10)})
```

```python
for col in df.columns:
    eksik_deger_sayisi = df[col].isnull().sum()
    print("{:<30}: {:>5} adet eksik değer bulunmaktadır.".format(col, eksik_deger_sayisi))

```

```python
df['name'].fillna('unknown', inplace=True)
df['host_name'].fillna('unknown', inplace=True)

# Tarihleri datetime formatına çevirme
df['last_review'] = pd.to_datetime(df['last_review'])

# Tüm tarihlerin ortalamasını alıyoruz
tarih_ortalama = df['last_review'].mean()

# Eksik tarih değerlerini ortalamayla doldurma
df['last_review'].fillna(tarih_ortalama, inplace=True)
df['reviews_per_month'].fillna(df['reviews_per_month'].mean(), inplace=True)

for col in df.columns:
    eksik_deger_sayisi = df[col].isnull().sum()
    print("{:<30}: {:>5} adet eksik değer bulunmaktadır.".format(col, eksik_deger_sayisi))

```

```python
# Fiyatı 0 olan değerleri eleyerek güncellenmiş veri setini oluşturma
df = df[df['price'] > 0]

# Güncellenmiş veri setinin boyutunu kontrol edelim
print("Güncellenmiş veri setinin boyutu:", df.shape)
```

Sayısal Değerlerin Dağılımı: Histogram

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['price'], kde=False, bins=30)
plt.xlabel('Fiyat')
plt.ylabel('Sayı')
plt.title('Fiyat Dağılımı')
plt.xlim(0, 10000)  # Fiyat ekseni 0 ile 1.000 arasında olacak
plt.show()
# nerede problem var?
```

Kategorik Değerlerin Dağılımı: Countplot

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='neighbourhood_group', data=df)
plt.xlabel('Mahalle Grupları')
plt.ylabel('Sayı')
plt.title('Mahalle Gruplarına Göre Konaklama Sayısı')
plt.xticks(rotation=45)
plt.show()
```

İlişki Gösterme: Scatter Plot

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='latitude', y='longitude', data=df, hue='neighbourhood_group', alpha=0.5)
plt.xlabel('Enlem')
plt.ylabel('Boylam')
plt.title('Mahalle Gruplarına Göre Enlem ve Boylam Dağılımı')
plt.legend(title='Mahalle Grupları', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

Kategorik Değerlerin Ortlama Fiyata Etkisi: Bar Plot

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='room_type', y='price', data=df)
plt.xlabel('Oda Türü')
plt.ylabel('Ortalama Fiyat')
plt.title('Oda Türlerine Göre Ortalama Fiyat')
plt.show()
```

    Korelasyon Matrisi: Heatmap

```python
plt.figure(figsize=(10, 8))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Korelasyon Matrisi')
plt.show()
```
